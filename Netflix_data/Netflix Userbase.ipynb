```python
# Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
import seaborn as sns
import os 
import json
import requests
```

```python
# Load new CSV file
netflix_userbase = Path("../csv_folder/Netflix_Userbase.csv")

# Read the CSV file into a DataFrame
netflix_userbase_df = pd.read_csv(netflix_userbase)
netflix_userbase_df.head()
```

```python
# Drop the plan duration column 
netflix_userbase_df.drop(columns=['Plan Duration'], inplace=True)
netflix_userbase_df.head()
```

```python
# Check for NaN values
print("NaN count per column:")
print(netflix_userbase_df.isnull().sum())

# Get rows with NaN
rows_with_nan = netflix_userbase_df[netflix_userbase_df.isnull().any(axis=1)]
print("\nRows with NaN values:")
print(rows_with_nan)
```

```python
# Save as a new CSV file
netflix_userbase_df.to_csv("../Netflix_data/Netflix_Userbase.csv", index=False)
```
